```python
# Sample corpus: 20 short sentences
texts = [
    "The cat sat on the mat.", "Dogs are loyal animals.", "She loves her dog.",
    "The sky is blue and beautiful.", "He loves to play football.",
    "This book is amazing.", "We watched a movie last night.",
    "My phone battery is dead.", "He is my best friend.", "That was a fantastic performance.",
    "Birds fly in the sky.", "She went to the market.", "Reading improves the mind.",
    "They went hiking in the mountains.", "The food was delicious.",
    "He bought a new laptop.", "The weather is cold today.",
    "Programming is fun.", "I like watching cricket.", "Nature is beautiful."
]
labels = [1]*10 + [0]*10
```

BoW Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(texts)

print("BoW matrix shape:", bow_matrix.shape)  # (20 samples Ã— N features)
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

 Vectorization (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
y = labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

 Train Naive Bayes & Logistic Regression

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

nb_model = MultinomialNB()
lr_model = LogisticRegression()

nb_model.fit(X_train, y_train)
lr_model.fit(X_train, y_train)
```

Evaluate with Accuracy, F1-Score, Confusion Matrix

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# Predictions
nb_preds = nb_model.predict(X_test)
lr_preds = lr_model.predict(X_test)

# Metrics
def evaluate(model_name, y_true, y_pred):
    print(f"\nðŸ“Š {model_name} Evaluation")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

evaluate("Naive Bayes", y_test, nb_preds)
evaluate("Logistic Regression", y_test, lr_preds)
```
