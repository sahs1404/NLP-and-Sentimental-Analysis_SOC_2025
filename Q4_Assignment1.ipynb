Load Passage and Split into 5 Segments

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import numpy as np

# Example: Replace with your Harry Potter passage
text = """
Harry looked around the dark hallway nervously. He could hear something moving in the shadows. 
A cold breeze passed by, making him shiver. Just then, a faint light appeared at the end of the hall. 
Hermione called out softly, “Harry, are you there?” He ran toward the voice, hopeful and scared.

Suddenly, the lights flared up and the hallway disappeared. He found himself in a beautiful meadow, 
sunlight streaming through the trees. The fear melted away as he heard laughter in the distance. 
He saw Ron waving at him from a hilltop. Relief washed over Harry as he ran toward his friends.

But as he got closer, the sky darkened. Clouds rolled in and the laughter turned into screaming. 
A storm broke out, thunder crashing around them. The meadow began to disappear into black fog. 
Harry screamed for Ron and Hermione, but they faded from sight.

The storm ended as suddenly as it started. Harry was alone in silence. A single feather floated down, 
glowing slightly. It landed on his hand, and he felt a strange warmth and calmness spreading. 
He took a deep breath and stood up, ready to move forward.
"""

# Split into 5 roughly equal segments
sentences = text.strip().split('\n')
segments = [' '.join(sentences[i::5]) for i in range(5)]
segments
```

Analyze Sentiment for Each Segment

```python
sid = SentimentIntensityAnalyzer()
sentiment_scores = [sid.polarity_scores(seg)['compound'] for seg in segments]
```

Plot Sentiment Trajectory

```python
plt.figure(figsize=(8,4))
plt.plot(range(1, 6), sentiment_scores, marker='o', linestyle='-', color='blue')
plt.title("Emotional Trajectory of the Passage")
plt.xlabel("Segment Number")
plt.ylabel("Sentiment Score (Compound)")
plt.axhline(0, color='gray', linestyle='--')
plt.grid(True)
plt.show()
```
